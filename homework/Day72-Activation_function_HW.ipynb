# Rectified Linear Unit- Relu 

f(x)=max(0,x)


```python
import numpy as np
from numpy import *
import matplotlib.pylab as plt
%matplotlib inline

'''
作業:
    寫出 ReLU & dReLU 一階導數
    並列印
'''
def ReLU(x):
    tmp = x.copy()
    tmp[tmp<0] = 0
    return tmp

def dReLU(x):
    tmp = x.copy()
    tmp[tmp >= 0] = 1
    tmp[tmp < 0] = 0
    return tmp
  
x = plt.linspace(-5,5,100)

#resize the X and Y axes
plt.gca().xaxis.set_major_locator(plt.MultipleLocator(1))
#plt.gca().yaxis.set_major_locator(plt.MultipleLocator(1))

#列印所有ReLU 值並輸出成一陣列
print(ReLU(x))
plt.plot(x, ReLU(x), 'r')
plt.show()
```
